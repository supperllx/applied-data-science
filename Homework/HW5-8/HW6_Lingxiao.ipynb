## 5. Distribution Plots

Using the distplot() method in the seaborn library to plot the distributions of normalized histograms for median income and normalized mean housing value in a single figure similar to the one shown below (you don't need to make your figure exactly same as this figure).
<img src="distplot.png">

````python
### BEGIN SOLUTION
income = housing['median_income']
value = housing['median_house_value']
normalized_income=(income-income.min())/(income.max()-income.min())
normalized_value = (value-value.min())/(value.max()-value.min())
fig, ax = plt.subplots() 
for data in [normalized_income, normalized_value]: 
    sns.distplot(data, ax=ax, bins=20, kde=True, label=data.name) 
    #ax.legend()
plt.xlabel('normalized median income & house value')
plt.legend()
### END SOLUTION
````

## 8. Visualize Geographical Data (10 points)

When a dataset contains geographical information (latitude and longitude), creating a geographical scatterplot of the data can reveal very useful information. 

a. Create a scatterplot using the __(latitude, longitude)__ as the cordinates, 1/50 of the population as the size of the marker, and the median house value as the color. You can use the "jet" colormap.

````python
fig,ax = plt.subplots(figsize=(10,6))
ax = housing.plot(kind='scatter', x='longitude', y='latitude', c=housing['median_house_value'],
                 s=(housing['population'])/50, alpha=0.4, colormap='jet', colorbar=True,ax=ax )
````

b. You can overlay the scatterplot on top of a map of the region. One simplest way to achieve this overlay is to load an map image and then show the image on the same axes which contains the scatterplot. You can pass an argument "zorder=0" to axes.imshow() and "zorder=1" to the scattplot method you have used.

You can use the image "ca.png" for the ca map. To read and show an image, you can use the following method:
```
img = mpt.image.imread("ca.png")
plt.imshow(img, extent=bounded_box)
```

````python
bounded_box = (housing.longitude.min(), housing.longitude.max(), housing.latitude.min(), housing.latitude.max())
### BEGIN SOLUTION
fig, ax = plt.subplots(figsize=(10,8))
img = mpt.image.imread('ca.png')
plt.imshow(img,extent=bounded_box,zorder=0)
ax = housing.plot(kind='scatter', x='longitude', y='latitude', c=housing['median_house_value'],
                 s=(housing['population'])/50, alpha=0.4, colormap='jet', colorbar=True,ax=ax ,zorder=1)
### END SOLUTION
````
